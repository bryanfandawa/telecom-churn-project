```python
df = df.drop(columns = ["customerID"])
```

```python
import re
df.columns = df.columns.str.replace(r'\xa0', ' ', regex = True).str.strip()
df.columns = df.columns.str.replace(r'(?<!^)(?=[A-Z])', '_', regex = True)
df.columns = df.columns.str.lower().str.replace(r'_t_v', '_tv', regex=True)
```

```python
df.columns.tolist()
```

```python
col_names = df.columns.tolist()
```

```python
#df = df.applymap(lambda x: str(x).strip(" '") if isinstance(x, str) else x)
```

```python
for i in col_names:
  print(f"----{i}----")
  print(df[i].unique())
  print("")
```

```python
col_yes_no = ["partner", "dependents", "paperless_billing", "churn", "phone_service"]
for i in col_yes_no:
  df[i] = df[i].map({"Yes": 1, "No": 0})
```

```python
col_no_internet = ["online_security", "online_backup", "device_protection", "tech_support", "streaming_tv", "streaming_movies"]
for i in col_no_internet:
  df[i] = df[i].map({"No internet service":0, "Yes":1, "No":0})
df["multiple_lines"] = df["multiple_lines"].map({"Yes": 1, "No": 0, "No phone service": 0})
```

```python
df = pd.get_dummies(df, columns=["internet_service", "contract", "payment_method"], drop_first=False, dtype=int)
```

```python
df["gender"] = df["gender"].map({"Male":1, "Female":0})
```

```python
df.head()
```

```python
pd.to_numeric(df["total_charges"], errors = "coerce").notna().all()
```

```python
mask = pd.to_numeric(df["total_charges"], errors = "coerce").isna()
df.loc[mask, "total_charges"]
```

```python
df["total_charges"] = df["total_charges"].astype(str).str.strip()
df["total_charges"].replace('', np.nan, inplace = True)
```

When you want to convert a string to something that's not a string, don't use the pandas function str.replace, use the built in python method .replace

```python
df.loc[mask,"total_charges"]
```

```python
df["total_charges"] = pd.to_numeric(df["total_charges"], errors = "coerce")
```

```python
df["total_charges"].isna().sum()
```

```python
df.dropna(subset = ["total_charges"], inplace = True)
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(r"\s+", "_", regex = True)
df.columns = df.columns.str.replace("(", "", regex = False)
df.columns = df.columns.str.replace(")", "", regex = False)
df.columns.tolist()
```

```python
df.to_csv("/content/cleaned_telecom_churn.csv", index=False)
```
